```python
from sklearn.ensemble import RandomForestRegressor
def set_miss_age(dataSet):
    df = dataSet[['Age', 'Pclass', 'SibSp', 'Parch', 'Fare']]
    
    know = df[df.Age.notnull()].values
    unknow = df[df.Age.isnull()].values
    
    x = know[:,1:]
    y = know[:,0]
    
    rft = RandomForestRegressor(random_state=12, n_jobs=1, n_estimators=10)
    rft.fit(x, y)
    
    preAge = rft.predict(unknow[:,1:])
    dataSet.loc[(df.Age.isnull()), 'Age'] = preAge
    
    return dataSet,rft
```

```python
def set_Cabin_type(dataSet):
    dataSet.loc[(dataSet.Cabin.notnull()), 'Cabin'] = 'No'
    dataSet.loc[(dataSet.Cabin.isnull()), 'Cabin'] = 'Yes'
    return dataSet
```

```python
dataSet, rft = set_miss_age(data_train)
```

```python
dataSet = set_Cabin_type(dataSet)
```

```python
dummies_Pclass = pd.get_dummies(dataSet.Pclass, prefix='Pclass')
dummies_Cabin = pd.get_dummies(dataSet.Cabin, prefix='Cabin')
dummies_Embarked = pd.get_dummies(dataSet.Embarked, prefix='Embarked')
dummies_Sex = pd.get_dummies(dataSet.Sex, prefix='Sex')
df = dataSet.join([dummies_Cabin, dummies_Embarked, dummies_Pclass, dummies_Sex])
df.drop([ 'Pclass', 'Cabin', 'Embarked', 'Sex', 'Name', 'Ticket'], axis=1, inplace=True)
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['Age_sc'] = scaler.fit_transform(df[['Age']])
df['Fare_sc'] = scaler.fit_transform(df[['Fare']])
df_scaler_train = df.drop(['Age', 'Fare'], axis=1, inplace=False)
df_scaler_train.head(10)
```

```python
from sklearn.linear_model import LogisticRegression
df_train = df_scaler_train.filter(regex='Survived|Age_.*|SibSp|Parch|Fare_.*|Cabin_.*|Embarked_.*|Sex_.*|Pclass_.*')
y = df_train.values[:, 0]
x = df_train.values[:, 1:]
clf = LogisticRegression()
clf.fit(x, y)
```
