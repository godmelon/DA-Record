```python
import matplotlib.pyplot as plt
import numpy as np
```

### 散点图

```python
x = np.linspace(0, 2 * np.pi, 50)
y = np.random.random(50)

plt.scatter(x, y)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.scatter(x, np.sin(x))
plt.show()
```

```python
# 使用plot绘制
plt.figure(figsize=(8,4))
plt.plot(x, np.sin(x), 'o');
```

```python
x = np.random.randint(1, 100, 100)
y = np.random.rand(100)

# size表示这100个点的大小，随机产生100个数
size = np.random.rand(100) * 500

# color表示颜色，随机产生100个sequence of color
color = np.random.rand(100)

# 设置画布size
plt.rc('figure', figsize=(10, 6))

plt.scatter(x, y, s=size, c=color,
            alpha=0.7,  # 设置点透明度，between 0 to 1，越小越透明
            linewidths=7)   # 设置圆点边缘宽度

plt.colorbar();
```

### 柱状图、条形图

```python
size = 5
y = np.random.random(size=size)
x = np.arange(size)

plt.bar(x, y, width=0.5);
```

堆积柱状图

```python
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)

ind = np.arange(N)
width = 0.35

p1 = plt.bar(ind, menMeans, width, color='#d62728', label='Men')
p2 = plt.bar(ind, womenMeans, width, bottom=menMeans, label='Women')

plt.ylabel('Score')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend()
plt.show()
```

带误差的堆积柱状图

```python
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N) 
width = 0.35

p1 = plt.bar(ind, menMeans, width, color='#d62728', yerr=menStd)
p2 = plt.bar(ind, womenMeans, width, bottom=menMeans, yerr=womenStd)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))
# plt.legend()
plt.show()
```
