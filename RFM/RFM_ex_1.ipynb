### 简易方法，通过pivot_table（类似透视表来实现）

```python
import pandas as pd
import numpy as np
import time

trad_flow = pd.read_csv('RFM_TRAD_FLOW.csv', encoding='gbk')
# 先将非标准字符串时间格式化为时间数组，再转换为时间戳便于计算
# trad_flow['time'] = trad_flow['time'].map(lambda x: time.strptime(x, '%d%b%y:%H:%M:%S'))
trad_flow.head()
```

```python
M = pd.pivot_table(trad_flow, index='cumid', columns='type', values='amount', aggfunc='mean')
M.head()
```

```python
F = pd.pivot_table(trad_flow, index='cumid', columns='type', values='transID', aggfunc='count', fill_value=0)
F.head()
```

```python
R = pd.pivot_table(trad_flow, index='cumid', columns='type', values='time', aggfunc='max', fill_value=None)
R.head(10)
```

```python
M['Special_offer'] = M['Special_offer'].fillna(0)
```

```python
M['spe_ratio'] = M['Special_offer'] / (M['Special_offer'] + M['Normal'])
M_rank = M.sort_values('spe_ratio', ascending=False, na_position='last')
M_rank.head(10)
```
